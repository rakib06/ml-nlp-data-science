```python
# case 1:  dependent variable: Genevia, independent variable: Chur
# x = Chur,  y =Genevia, , z = stGallen
from numpy import arange
error = 1;
def error(actual_value, exp_value):
    return abs(exp_value-actual_value)/actual_value
actual_x = 118
actual_y = 290
min_error = 1
def cal_y(x):
    return x-36+200;
for x in arange(110, 130, .05):
    y = cal_y(x)
    total_error = error(actual_x, x) + error(actual_y,y)
    # print(error(actual_x, x),  error(actual_y,y), "y =", y)
    if total_error < min_error:
        print("--------"*10)
        print(x, '  ==>', total_error)
        min_error = total_error
        
        print(error(actual_x, x),  error(actual_y,y), " x= ",x, " y =", y, " z=",x-210 )
        
print(min_error)


```
